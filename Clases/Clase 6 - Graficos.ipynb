**Ejercicio 1**
Construir 3 gráficos de línea a partir de un array creado con la función linspace.

a. Uno que represente una curva lineal. <br>
b. Otro que represente una curva exponencial. <br>
c. Un tercero que represente una curva logarítmica.<br>

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Uno que represente una curva lineal.

N=8
x = np.linspace(0,6,N)

plt.plot(x)
```

```python
# Otra forma
z= np.linspace(0,50,100)
plt.plot(z,2 * z)
```

```python
# Otro que represente una curva exponencial.

y = x ** 2
plt.plot(x,y)
```

```python
# Un tercero que represente una curva logarítmica.
r= np.linspace(1,20,10)
y = np.log(r)
plt.plot(r,y)
```

```python
N=8
x = np.linspace(0,6,N)
plt.plot(x, np.exp(x), "b", label = "Exponencial")
plt.plot(x, 3*x, ":r", label = "Lineal y=3X")
plt.title("Grafico")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.legend()
```

```python
import pandas as pd
df= pd.read_csv("../Recursos/500_Person_Gender_Height_Weight_Index.csv", delimiter=";")
df
```

```python
#constuir un gráfico de dispersión (scatter), comparando altura y peso y 
# diferenciando por color el genero. 
# ¿Qué conclusiones se pueden obtener a partir de este gráfico?
df["Gender"] = df["Gender"].apply(lambda x: "r" if x == "Male" else "g")
plt.scatter(df["Weight"], df["Height"],  c=df["Gender"])
```

```python
# Utilizando el dataset "StudentsPerformance.csv", 
# comparar la distribución de notas de matemáticas y de lectura. 
# Utilizar ambos histogramas sobre una misma figura. 
# ¿Qué conclusiones podemos sacar sobre estos resultados?

students= pd.read_csv("../Recursos/StudentsPerformance.csv")
students
```

```python
plt.hist((students["reading score"],students["math score"]), bins=20,label=["Reading", "Math"])
plt.title("Histograma de notas")
plt.legend()
```

```python
# Comparar el promedio de notas entre los alumnos que completaron el test de preparación
# y los que no. ¿Qué conclusiones podemos sacar sobre estos resultados?

#Usamos un grafico de barras para agrupar los datos.

students.groupby("test preparation course").mean().plot(kind="bar", y="reading score", ylabel="Promedio de notas", title="Comparativo de promedios")
```

```python
# Construir un gráfico de torta que muestre la proporción de aprobados en matemáticas
# entre hombres y mujeres. Se aprueba con 60.
students[students["math score"]>=60].groupby("gender").size().plot(kind="pie", title="Aprobados en matematicas")
```

```python
# Construir un gráfico de columnas basado en el dataset "Listado_ventas.xlsx" 
# que muestre la evolución de ventas por año y mes.

listado_ventas= pd.read_excel("../Recursos/Listado_ventas.xlsx")
listado_ventas
```

```python
#Convierto a tipo de dato fecha
listado_ventas["FECHA VENTA"] = pd.to_datetime(listado_ventas["FECHA VENTA"]) 
#Agrego una columna de año-mes
listado_ventas["AÑO-MES"] = listado_ventas["FECHA VENTA"].dt.strftime('%Y-%m') 

valores = listado_ventas[["AÑO-MES","MONTO"]].groupby("AÑO-MES").mean().plot(kind="bar")
```

## Laboratorio: Matplotlib 5

Utilizar el dataset "Listado_ventas.xlsx"para  Construir una figura que incluya dos gráficos:
-  Suma de montos de ventas por sección.
- Suma de montos de ventas por vendedor.

```python
plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
resumen = listado_ventas.groupby("SECCIÓN").sum()
plt.pie(resumen["MONTO"], labels = resumen.index, autopct='%1.1f%%' )
plt.subplot(1,2,2)
resumen = listado_ventas.groupby("VENDEDOR").sum()
plt.bar(height = resumen["MONTO"], x = resumen.index )
```

```python
# Con orientacion a objetos
figs, axs =  plt.subplots(1,2)

resumen = listado_ventas.groupby("SECCIÓN").sum()
axs[0].pie(resumen["MONTO"], labels = resumen.index, autopct='%1.1f%%' )
resumen = listado_ventas.groupby("VENDEDOR").sum()
axs[1].bar(height = resumen["MONTO"], x = resumen.index)
axs[0].set_title('Ventas por sección') 
axs[1].set_title('Ventas por vendedor')

#Establece el tamaño
figs.set_size_inches(15,6) 
figs.set_facecolor("g") #Establece color de fondo
```
