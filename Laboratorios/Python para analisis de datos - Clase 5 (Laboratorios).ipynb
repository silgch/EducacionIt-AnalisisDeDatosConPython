## Modulo 5 -  Laboratorio Pandas VI

Importar el data set “titanic.csv” y calcular: promedio de edad, edad máxima, edad mínima y cantidad, todo agrupado por sobrevivientes y no sobrevivientes

```python
import pandas as pd
data = pd.read_csv("../Recursos/titanic.csv")
print("El promedio de la edad por sobrevivientes es: \n", data[["Age","Survived"]].groupby("Survived").mean())
```

```python
print("La edad maxima de los  sobrevivientes es: \n", data[["Age","Survived"]].groupby("Survived").max())
```

```python
print("La edad minima  de los  sobrevivientes es: \n", data[["Age","Survived"]].groupby("Survived").min())
```

```python
print("La cantidad de sobrevivientes fue: \n", data.groupby("Survived").size())
```

```python
#O bien podemos calcular todo junto(para la edad):
data[["Survived","Age"]].groupby("Survived").agg(["mean","min","max","count"])
```

## Modulo 5 -  Laboratorio Pandas VII
Importar el data set “titanic.csv” y realizar los siguientes gráficos:
1. Gráfico de columnas que muestre la cantidad de pasajeros por grupo etario (10 en 10).
2. Gráfico de columnas que muestre la cantidad de sobrevivientes por grupo etario (10 en 10).
3. Gráfico de columnas que muestre la cantidad de no sobrevivientes por grupo etario (10 en 10).

```python
#Gráfico de columnas que muestre la cantidad de pasajeros por grupo etario (10 en 10).
data.groupby(data["Age"]//10).size().plot.bar(title = "Cantidad de pasajeros por grupo etario") 
```

```python
#Gráfico de columnas que muestre la cantidad de sobrevivientes por grupo etario (10 en 10).
data.groupby(data.loc[data["Survived"] == 1, "Age"]//10).size().plot.bar(title="Cantidad de sobrevivientes por grupo etario")
```

```python
# Gráfico de columnas que muestre la cantidad de no sobrevivientes por grupo etario (10 en 10).
data.groupby(data.loc[data["Survived"] == 0, "Age"]//10).size().plot.bar(title="Cantidad de no sobrevivientes por grupo etario")
```

2. Calcular la suma de monto de ventas y la cantidad de operaciones agrupado por cliente.

```python
ventas = pd.read_excel("../Recursos/Ventas.xlsx")
```

```python
ventas
```

```python
# Para los valores nulos de ventas, poner el valor del promedio de ventas totales.
ventas["Monto Venta"]= ventas["Monto Venta"].fillna(ventas["Monto Venta"].mean())
```

```python
ventas
```

```python
# Calcular la suma de monto de ventas y la cantidad de operaciones agrupado por cliente.
ventas[["Cliente","Monto Venta"]].groupby("Cliente").agg(["sum","count"])
```

## Modulo 5 - Laboratorio Pandas X
Importar el archivo "GoldPrice.csv" y realizar los siguientes cálculos:
1. Determinar el promedio de precios por año.
2. Calcular el año con mayor promedio y el año con menor promedio.
Devolver los resultados bajo el formato:<br>
    -Mayor precio promedio<br>
        -Año:xxx<br>
        -Precio:xxxx<br>
    -Menor precio promedio<br>
        -Año:xxxx<br>
        -Precio: xxxx<br>

```python
#Determinar el promedio de precios por año.
gold= pd.read_csv("../Recursos/GoldPrice.csv")
gold["Date"]=pd.to_datetime(gold["Date"])
gold["Año"] = gold["Date"].dt.year
gold[["Año","Price"]].groupby("Año").mean()
```

```python
# Calcular el año con mayor promedio y el año con menor promedio.
#Año máximo promedio
año = int(gold[["Año","Price"]].groupby("Año").mean().idxmax())
precio = float(gold[["Año","Price"]].groupby("Año").mean().max())
print("Año: ", año)
print("Precio: ", precio)
```

```python
#Año minimo promedio
año = int(gold[["Año","Price"]].groupby("Año").mean().idxmin())
precio = float(gold[["Año","Price"]].groupby("Año").mean().min())
print("Año: ", año)
print("Precio: ", precio)
```
