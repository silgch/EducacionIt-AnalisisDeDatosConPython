```python
# Crear series a partir de listas y arrays unidimensionales.
import pandas as pd
import numpy as np

paises= pd.Series(["Argentina","Venezuela","Chile"])
print(paises)

monedas= pd.Series(np.array(["Peso", "Bolivar","Peso chileno"]))
print(monedas)
```

```python
# Crear series especificando las etiquetas del índice.
pd.Series([106,1,200], index=["Argentina","Venezuela","Chile"], dtype="float32",name="cotizacion_divisas")
```

```python
# Crear dataframes a partir de :

    # listas anidadas
edades= pd.DataFrame([["Luis", 30],["Mary",32],["Vanesa", 45]], columns= ["Empleado","Edad"]) 
print(edades)
    # arrays bidimensionales 
numeros= pd.DataFrame(np.array([[1,2,3],[4,5,6]]))
print(numeros)
    # diccionarios.
calorias= pd.DataFrame({"Producto": ["Edulcorante",  "Coca-light", "Azucar", "Agua"], "Calorias": [0,0,250,0]})
print(calorias)
```

```python
# Crear dataframes especificando etiquetas para filas y columnas.
productos= pd.DataFrame([["Campera","Pantalon","Buzo"], ["Adidas","Nike","Topper"],[12000,8500,4500]], columns=["Producto","Marca","Precio"], index=[1001,1002,1003])
print(productos)
```
