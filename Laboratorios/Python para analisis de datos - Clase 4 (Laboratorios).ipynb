## Módulo 04 - Laboratorio 04
1. Crear una columna llamada “SurvivalAge” que sea el producto de la columnas “Survived” y “Age”. Hacerlo por operador y por método.<br>
2. Crear una nueva columna que genere un rango etario para cada persona, de modo que las edades entre 0 y 9 generen el número 1, entre 10 y 19 el número 2, etc. de modo que los missing values se mapeen a cero.<br>
3. Seleccionar los datos de todas las mujeres de primera clase.<br>
4. Seleccionar los datos de los varones menores de edad de segunda clase.

```python
# Crear una columna llamada “SurvivalAge” que sea el producto de la columnas “Survived” y “Age”. Hacerlo por operador y por método.
df["SurvivalAge"]=df["Survived"] * df["Age"]
print(df["SurvivalAge"])
```

```python
# Crear una nueva columna que genere un rango etario para cada persona, 
# de modo que las edades entre 0 y 9 generen el número 1, 
# entre 10 y 19 el número 2, etc. de modo que los missing values se mapeen a cero.




df["Rango Etario"] = np.ceil(df["Age"]/10)
#df["Rango Etario"]= np.ceil(df["Rango Etario"])

print(df.loc[(df["Age"] < 20 )& (df["Age"] > 9 ), ["Age", "Rango Etario"]] )
```

```python
# Seleccionar los datos de todas las mujeres de primera clase.
df[(df["Sex"] == "female" )& (df["Pclass"] == 1 )]
```

```python
# Seleccionar los datos de los varones menores de edad de segunda clase.
df[(df["Sex"] == "male" )& (df["Pclass"] == 2) &( df["Age"] < 18)]
```

## Módulo 04 - Laboratorio 05
1. Encontrar el promedio de edad de hombres y mujeres.
2. Encontrar la cantidad de hombres y mujeres.
3. Encontrar la cantidad de supervivientes entre hombres y mujeres.
4. Encontrar la proporción de supervivientes entre hombres y mujeres.
5. Usar describe con incrementos del percentil de 10%.

```python
#Encontrar el promedio de edad de hombres y mujeres.

print("El promedio de la edad de los hombres es: ", df[(df["Sex"]=="male")]["Age"].mean())
```

```python
#Usando LOC:
print("El promedio de la edad de las mujeres es: ", df.loc[(df["Sex"]=="female"),"Age"].mean())
```

```python
#Encontrar la cantidad de hombres y mujeres.
print("La cantidad de hombres es: ",df.loc[(df["Sex"]=="male"),"ID"].count() )
print("La cantidad de mujeres es: ",df.loc[(df["Sex"]=="female"),"ID"].count() )
```

```python
#Encontrar la cantidad de supervivientes entre hombres y mujeres.
supervivientes= df[df["Survived"]== 1]
print("La cantidad de supervivientes es: ",supervivientes["ID"].count() )
print("La cantidad de mujeres supervivientes es: ",supervivientes.loc[(supervivientes["Sex"]=="female"),"ID"].count() )
print("La cantidad de hombres supervivientes es: ",supervivientes.loc[(supervivientes["Sex"]=="male"),"ID"].count() )
```

```python
#Tambien podemos hacer el mismo calculo de forma mas simple usando value_counts:
supervivientes["Sex"].value_counts()
```

```python
# Encontrar la proporción de supervivientes entre hombres y mujeres.
supervivientes["Sex"].value_counts(normalize=True)
```

```python
# Usar describe con incrementos del percentil de 10%.
df.describe(percentiles= [.1, .2, .3,.4,.5,.6,.7,.8,.9])
```
