## Matplotlib 1

La visualización de datos es la representación de los datos a través de gráficos de distinto tipo. Es un tipo de representación mucho más fácil de entender para una persona que otras formas de representar información, como las tablas, más aún cuando la cantidad de datos es grande.
<br>
<br>
Las representaciones visuales permiten descubrir y comunicar relaciones complejas entre los datos y son una herramienta importante del análisis de datosmoderno.

Matplotlib es un módulo de bajo nivel para visualización de datos. Está hecho sobre Numpy,
por lo que es ideal para hacer gráficos a partir de datos alojados en arrays. Por esta razón se integra perfectamente al ecosistema de módulos de Python. <br><br>
Se dice que es de bajo nivel porque permite controlar cualquier aspecto de un gráfico, ya que
tiene numerosas funciones para configurar cualquier cosa que uno quiera. Por esto mismo sirve de base a otros módulos (de alto nivel) que están construidos sobre Matplotlib, como Seaborn o Cartopy.

## Interfaces
Matplotlib ofrece dos interfaces (o formas de interactuar con las funcionalidades):
<br>
● La interfaz “pyplot”.<br>
● La interfaz orientada a objetos.<br>
<br>
Las dos interfaces dan acceso a la misma funcionalidad, sin embargo, la primera es más
cómoda a la hora de realizar gráficos sencillos de forma rápida mientras que la segunda es
algo más conveniente a la hora de hacer gráficos complejos.

## Plot

defecto una línea continua azul).** Recibe como parámetros dos arrays, uno para las coordenadas x y
otro para las coordenadas y de los datos.<br>
En caso de que se provea un solo array, este se usa como coordenadas y mientras que las coordenadas x se
defaultean a números enteros empezando por cero.

```python
N=8
x = np.linspace(0,6,N)
y= np.sin(x)
plt.plot(x,y)
```

## Matplotlib 2

### Format string
Adicionalmente a los datos, a la función plot se le puede pasar un string para controlar algunos aspectos básicos del gráfico de forma rápida y sencilla. Este format string puede contener información para controlar tres aspectos:
<br><br>
● Marcador: controla el tipo de marca que se usa para señalar los datos, como por ejemplo si los vamos a marcar con círculos, cuadrados, triángulos, puntos, estrellas, etc.<br>
● Línea: el tipo de línea a utilizar para unir los puntos: continua, punteada, rayada, raya-punto o ninguna.<br>
● Color: algunos colores básicos que podemos usar para el conjunto de datos<br>

### Marker
Cada opción tiene un carácter que lo representa y puede estar presente o no en el **format** string.<br>
Las opciones para el marcador son:<br>


### Línea y color
Los caracteres para el tipo de línea y para el color se muestran a continuación. Si el carácter de la línea se omite, solo graficará los puntos.


### Ejemplos


```python
# s --> Markers: square marker
# : --> Line Style: dashed line style
# r --> Line Style: color red

plt.plot([1,5,7,3,6,9,3], "s:r")
```

```python
# ^ --> Markers: triangle_up marker
# -- --> Line Style: dotted line style
# g --> Line Style: color green

plt.plot([1,5,7,3,6,9,3], "^--g")
```

```python
# o --> Markers: circle marker
# b --> Line Style: color blue
plt.plot([1,5,7,3,6,9,3], "ob")
```

```python
# : --> Line Style: dotted line style
# k --> Line Style: color black
plt.plot([1,5,7,3,6,9,3], ":k")
```

### Argumentos
Si bien es posible controlar algunos aspectos básicos con el **format string**, para un control más completo de los aspectos del gráfico debemos usar los parámetros de la función.

```python
plt.plot([1,5,6,3,8], marker="s", markersize=15, markerfacecolor= "red", markeredgecolor= "black", markeredgewidth=3,linewidth=2 )
```

## Matplotlib 3

### Anatomía de un gráfico
Un gráfico de matplotlib se puede componer de muchas partes, cada una representada por un tipo de objeto particular, constituyendo una jerarquía de objetos.<br>
El más grande de ellos es la figura, que puede contener uno o más subgráficos y estos a su vez pueden contener otros objetos como líneas, puntos, leyendas, etc.

Para controlar las propiedades de la figura se usa la función **figure.** Esta función crea una nueva figura.
Cuando usamos solamente la función plot la figura se crea implícitamente con los valores por defecto.
Al crearla explícitamente podemos controlar sus parámetros, como por ejemplo el tamaño de la figura.
<br>
El parámetro **figsize recibe una tupla de números con las dimensiones de ancho y alto total** de la figura medido en pulgadas.

### Figura activa
La interfaz de pyplot trabaja manteniendo un registro de cuál es la figura activa, de modo que
las funciones que se invocan actúan sobre ella.<br>
Por ejemplo, si llamamos a la función plot muchas veces seguidas actúan sobre el mismo gráfico. Para que actúen sobre figuras distintas hay que crear un nueva figura mediante figure.<br>
Así podemos representar varios conjuntos de datos en una o varias figuras.

```python
x = np.array([0,1,2,3,4,5])
plt.plot(x, "o-r")
plt.plot(x**2, "s:g")
```

```python
x = np.array([0,1,2,3,4,5])
plt.plot(x, "o-r")
plt.figure()
plt.plot(x**2, "s:g")
```

### Otros elementos del gráfico
Del mismo modo podemos agregar más elementos siempre teniendo en cuenta que las funciones actúan
sobre la figura activa. Por ejemplo la función **title** permite agregarle un título a la figura. Las funciones **xlabel e ylabel** permiten definir el texto de los ejes.<br>
Si cada conjunto de datos tiene definido un label, la función **legend** permite activar el recuadro de leyendas. El label se define dentro de la función que construye el gráfico con el parámetro del mismo nombre.<br>
**No importa el orden en que se vayan llamando a las distintas funciones.**

```python
x = np.array([0,1,2,3,4,5])
plt.plot(x, "o-r", label = "lineal")
plt.plot(x**2, "s:g", label="cuadratica")
plt.title("Grafico 1")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.legend()
```

### Más elementos
Se pueden agregar muchos elementos más llamando a las funciones adecuadas. Por ejemplo podemos
agregar textos en cualquier parte del gráfico con la **función text, líneas verticales u horizontales con axvline y axhline.** <br>
También **se puede activar una grilla con la función grid** o activar y desactivar las líneas pequeñas sobre los ejes con **minorticks_on / minorticks_off** <br>
En general, todas las funciones tienen numerosos parámetros para controlar cómo se muestra cada elemento.

## Matplotlib 4

### Otros tipos de gráficos
Vimos que el módulo pyplot tiene numerosas funciones para activar distintos elementos dentro
de un gráfico. También tiene varias funciones para realizar distintos tipos de gráficos.<br>
Hasta ahora solo vimos la función plot, pero encontramos también las funciones **scatter,
hist, bar, barh, pie, violinplot, boxplot, etc.**  Con ellas vamos a poder representar información de diversas formas.

### Gráfico de puntos
La función **scatter** función permite graficar puntos en el plano donde además podemos controlar el
color y el tamaño de cada punto para representar más información aparte de las coordenadas de los
puntos, como categorías o magnitudes continuas asociadas a cada punto.

```python
x, y = np.random.random((2,30))
c = np.random.randint(0,9,30)
s = np.random.randint(10,150,30)
```

```python
plt.scatter(x, y, c=c, s=s)
```

### Histogramas
Se pueden crear histogramas con la función **hist.**<br>
Este tipo de gráficos es útil para apreciar la **distribución de números.** Se conforma de **barras** cuyo alto es la cantidad de datos que están entre los valores determinados por el ancho de la barra.<br>
Se puede controlar la cantidad de barras con el parámetro **bins.** A su vez se puede hacer varios histogramas en una misma figura y controlar cómo se muestran juntas con el parámetro **histtype.**

```python
x1= np.random.normal(4,3,1000)
x2= np.random.normal(1,1,1000)
plt.hist((x1,x2), bins=20, label=["Grupo 1", "Grupo 2"])
plt.title("Histograma")
plt.legend()
```

```python
x1= np.random.normal(4,3,1000)
x2= np.random.normal(1,1,1000)
plt.hist((x1,x2), bins=30, histtype="stepfilled", alpha=0.5)
plt.title("Histograma")
```

### Violinplot y Boxplot
Otros gráficos útiles para ver la distribución de números son violinplot y boxplot. El primero muestra una distribución
**suavizada de los datos** y el segundo **marca la mediana y los cuartiles de la distribución de números.**


### Gráfico de barras
Este tipo de gráfico permite visualizar cantidades mediante barras en donde **la longitud de cada barra es proporcional a la magnitud que se quiere representar.**  Se suele utilizar para
representar información que pertenece a categorías distintas.<br>
La **función bar** necesita dos parámetros x e y que representan las posiciones de los centros de cada barra en el eje x y la altura de cada barra en el eje y respectivamente. Así, podemos poner barras en cualquier posición del eje x y con cualquier altura. El ancho de las barras las podemos controlar con el parámetro **width.**  Las etiquetas del eje x se controlan con **tick_label.** La función **barh permite hacer barras horizontales.**

```python
plt.bar([1,2,3],[18,25,47], tick_label=["A", "B","C"])
```

### Gráfico de tortas
Este tipo de gráfico es muy útil para mostrar las proporciones relativas entre distintas cantidades
visualizándolas como cuñas de un mismo círculo.<br>
Se realizan con la función pie que toma las magnitudes a representar.<br>
Se le puede especificar el parámetro **labels para asignar textos a cada cuña**. También está el
parámetro explode, que permite separar las cuñas del centro del gráfico para destacar alguna de ellas.

```python
plt.pie([12,18,35,20], labels=list("ABCD"))
plt.title("Grafico de torta")
```

```python
plt.pie([12,18,35,20], labels=list("ABCD"), explode=[0.4,0.2,0.1,0], shadow=True)
plt.title("Grafico de torta")
```

## Matplotlib 5

### Subplots
En ciertas situaciones podemos querer representar varios gráficos dentro de la misma figura. Podemos crear varios plots dentro de la misma figura con la función **subplot.**<br>
Esta función añade un **subgráfico** a la figura activa y matplotlib queda enfocado en ese subplot, de modo que las acciones que hagamos repercuten sobre ese subgráfico. Para pasar a otro subgráfico debemos llamar a subplot otra vez.<br>
La función subplot toma la cantidad de filas y columnas en las que deben acomodarse los subgráficos. Por ejemplo, dos filas y una columna conforman un gráfico sobre otro.<br>
Dos filas y dos columnas conforman cuatro subgráficos. El tercer parámetro es el índice del subgráfico sobre el que enfocarse.

```python
plt.figure(figsize=(8,3))
plt.subplot(1,2,1)
plt.hist(np.random.randn(500))
plt.title("Histograma")
plt.subplot(1,2,2)
plt.pie([25,15,44,16])
plt.title("Grafico de tortas")
```

```python
x = np.linspace(-10,10,400)
plt.figure(figsize=(6,4))
plt.subplot(2,1,1)
plt.plot(x,np.cos(x))
plt.title("Coseno")
plt.subplot(2,1,2)
plt.plot(x,np.sin(x))
plt.title("Seno")
```

### Interfaz orientada a objetos
Hasta ahora vimos que la interfaz pyplot mantiene un registro de cuál es la figura activa
y las funciones que ejecutamos tienen efecto sobre esa figura. Para que las funciones tengan
efecto en otra parte tenemos que activar una nueva figura o subplot.<br>
La interfaz orientada a objetos permite **ponerle nombre a los objetos como figuras y subplots,**
de modo que podamos referirnos a ellos en cualquier parte y en cualquier orden. Y ahora ya no vamos a trabajar con funciones sino con **métodos sobre esos objetos.**

### plt.subplots
Para crear estos objetos usamos la función **subplots** (en plural) que crea una figura y uno o más subplots.<br>
Debemos pasarle la **cantidad de filas y columnas que van a diagramar a los subplots** (por defecto 1 y 1, es decir, una figura con un subplot).

```python
fig, ax= plt.subplots()
```

```python
# Si tenemos más de un subplot el segundo elemento es un array que contiene a los objetos de cada subplot.

fig, axs= plt.subplots(2,2)
```

```python
print(type(axs), axs.shape)
```

### Axes
Los objetos de tipo axes tienen numerosos métodos para actuar sobre ellos. Esencialmente, toda función de pyplot tiene un método respectivo para los objetos de tipo axes. 

```python
fig, ax= plt.subplots()
ax.plot([1,2,4,8,16],label="Datos")
ax.set_title("Grafico subplots")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.legend()
```

### Subplots
Al hacer subgráficos tenemos que acceder a cada objeto del array usando los índices correspondientes.

```python
fig, axs= plt.subplots(2,2)
axs[0,0].plot([3,6,7,9])
axs[0,1].hist(np.random.randn(20))
axs[1,0].pie([3,5,6,7])
axs[1,1].bar([3,5,6,7],[1,2,3,4])
```
